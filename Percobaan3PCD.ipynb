```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/Fuji_apple.jpg')
img_real = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
smoothed_image = cv2.blur(img_real,(5, 5))

plt.subplot(1, 2, 1)
plt.title("original")
plt.imshow(img_real)

plt.subplot(1, 2, 2)
plt.title("Neighborhood Averaging")
plt.imshow(smoothed_image)
plt.show
```

```python
median_filtered = cv2.medianBlur(img_real, 9)

plt.subplot(1, 2, 1)
plt.title("original")
plt.imshow(img_real)

plt.subplot(1, 2, 2)
plt.title("Median Blur")
plt.imshow(median_filtered)
plt.show
```

```python
kernel = np.array([[0, -1,0],
                   [-1, 5, -1],
                   [0, -1, 0]])

# Sharpen the image
high_pass_filtered = cv2.filter2D(img_real, -1, kernel)

plt.subplot(1, 2, 1)
plt.title("original")
plt.imshow(img_real)

plt.subplot(1, 2, 2)
plt.title("High-pass Filtering")
plt.imshow(high_pass_filtered)
plt.show
```

```python
import cv2
import os
import matplotlib.pyplot as plt
from google.colab import files
```

```python
uploaded = files.upload()

uploaded_files = list(uploaded.keys())
input_image_path = uploaded_files[0]

output_image_path = "gambar_compressed.jpg"
```

```python
kualitas_kompresi = 60
image = cv2.imread(input_image_path)
cv2.imwrite(output_image_path, image [int(cv2.IMWRITE_JPEG_QUALITY), kualitas_kompresi])
```

```python
gambar_asli = cv2.imread(input_image_path)
gambar_kompresi = cv2.imread(output_image_path)

gambar_asli_rgb = cv2.cvtColor(gambar_asli, cv2.COLOR_BGR2RGB)
gambar_kompresi_rgb = cv2.cvtColor(gambar_kompresi, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Gambar Asli')
plt.imshow(gambar_asli_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Gambar Kompresi')
plt.imshow(gambar_asli_rgb)
plt.axis('off')

plt.show()
```

```python
ukuran_sebelum_kompresi = os.path.getsize(input_image_path)
ukuran_setelah_kompresi = os.path.getsize(output_image_path)

print(f"Ukuran Sebelum Kompresi: {ukuran_sebelum_kompresi} bytes")
print(f"Ukuran Setelah Kompresi: {ukuran_setelah_kompresi} bytes")
print(f"Rasio Kompresi: {ukuran_setelah_kompresi / ukuran_sebelum_kompresi * 100:.2f}%")
```

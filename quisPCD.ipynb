```python
import cv2
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = "/content/gambar.jpeg"
```

```python
def read_image_opencv(image_path):
    img = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img_rgb
```

```python
def read_image_pil(image_path):
    img = Image.open(image_path)
    if img.mode != 'RGB':
        img = img.convert('RGB')
    return img
```

```python
def sobel_edge_detection(image):
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=5)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=5)
    magnitude = np.sqrt(sobelx**2 + sobely**2)
    magnitude *= 255.0 / magnitude.max()
    return magnitude
```

```python
def canny_edge_detection(image):
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    edges = cv2.Canny(gray, 100, 200)
    return edges
```

```python
def random_crop(image, crop_height, crop_width):
    height, width = image.shape[:2]
    y = np.random.randint(0, height - crop_height)
    x = np.random.randint(0, width - crop_width)
    cropped = image[y:y+crop_height, x:x+crop_width]
    return cropped
```

```python
def invert_image(image):
    inverted_image = cv2.bitwise_not(image)
    return inverted_image
```

```python
def salt_and_pepper_noise(image, amount=0.05):
    row, col = image.shape[:2]
    s_vs_p = 0.5
    num_salt = np.ceil(amount * image.size * s_vs_p)
    num_pepper = np.ceil(amount * image.size * (1. - s_vs_p))
    coords_salt = [np.random.randint(0, i-1, int(num_salt)) for i in (row, col)]
    coords_pepper = [np.random.randint(0, i-1, int(num_pepper)) for i in (row, col)]
    image[coords_salt[0], coords_salt[1]] = 255
    image[coords_pepper[0], coords_pepper[1]] = 0
    return image
```

```python
def gaussian_blur(image, kernel_size=(5, 5), sigma_x=0):
    blurred_image = cv2.GaussianBlur(image, kernel_size, sigma_x)
    return blurred_image
```

```python
def rotate_image(image, angle):
    height, width = image.shape[:2]
    center = (width // 2, height // 2)
    matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated_image = cv2.warpAffine(image, matrix, (width, height))
    return rotated_image
```

```python
def flip_vertical(image):
    flipped_image = cv2.flip(image, 0)
    return flipped_image
```

```python
def flip_horizontal(image):
    flipped_image = cv2.flip(image, 1)
    return flipped_image
```
